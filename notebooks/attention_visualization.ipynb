```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('mean_attention.json', 'r', encoding='utf-8') as f:
    mean_data = json.load(f)
```

```python
with open('attention.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
```

```python
def visualize_one_layer(data, layer):
    attention = data['attention_e'][layer]
    tokens = [token.replace('Ġ', ' ') for token in data['tokens']]

    plt.figure(figsize=(12, 5))
    plt.bar(range(len(tokens)), attention, color='skyblue')
    plt.xticks(range(len(tokens)), tokens, rotation=45, ha='right')
    plt.ylabel('Attention to e')
    plt.title(f'Layer {layer}: attention to e-token')
    plt.tight_layout()
    plt.show()

def visualize_all_layers(data):
    attention = data['attention_e']
    tokens = [token.replace('Ġ', ' ') for token in data['tokens']]

    plt.figure(figsize=(14, 6))
    plt.imshow(attention, aspect='auto')
    plt.yticks(range(len(attention)), range(len(attention)))
    plt.xticks(range(len(tokens)), tokens, rotation=90)
    plt.xlabel('Token')
    plt.ylabel('Layer')
    plt.colorbar(label='Attention value')
    plt.title('Attention to e-token')
    plt.show()
```

```python
def visualize_one_head(data, layer, head):
    attention = data['attention_e'][layer][head]
    tokens = [token.replace('Ġ', ' ') for token in data['tokens']]

    plt.figure(figsize=(12, 5))
    plt.bar(range(len(tokens)), attention, color='skyblue')
    plt.xticks(range(len(tokens)), tokens, rotation=45, ha='right')
    plt.ylabel('Attention to e')
    plt.title(f'Layer {layer}, Head {head}: attention to e-token')
    plt.tight_layout()
    plt.show()

def visualize_all_heads(data, layer):
    attention = data['attention_e'][layer]
    tokens = [token.replace('Ġ', ' ') for token in data['tokens']]

    plt.figure(figsize=(14, 6))
    plt.imshow(attention, aspect='auto')
    plt.yticks(range(len(attention)), range(len(attention)))
    plt.xticks(range(len(tokens)), tokens, rotation=90)
    plt.xlabel('Token')
    plt.ylabel('Layer')
    plt.colorbar(label='Attention value')
    plt.title(f'Layer {layer} attention to e-token')
    plt.show()
```

```python
idx = 0
print(f'Text:\n{mean_data[idx]['text']}')
visualize_one_layer(mean_data[idx], 8)
visualize_all_layers(mean_data[idx])
```

```python
print(f'Text:\n{data[idx]['text']}')
visualize_one_head(data[idx], 8, 16)
visualize_all_heads(data[idx], 8)
```
