```python
import json
import matplotlib.pyplot as plt
```

```python
with open('D:\\учеба\\diploma\\data\\noise_effect.json', 'r') as f:
    data = json.load(f)
```

```python
def visualize_noises(data, noise_types):
    plt.figure()

    for noise_type in noise_types:
        alpha = []
        accuracy = []
        for item in data:
            if item['noise'] == noise_type:
                alpha.append(item['alpha'])
                accuracy.append(item['accuracy'])
        plt.plot(alpha, accuracy, marker='o', label=noise_type)

    plt.xlabel('alpha')
    plt.ylabel('accuracy')
    plt.title('Accuracy vs alpha for different noise types')
    plt.grid(True)
    plt.legend()
    plt.show()
```

```python
noise_types = ['gaussian', 'uniform', 'sinusoidal', 'exponential']
visualize_noises(data, noise_types)
```
