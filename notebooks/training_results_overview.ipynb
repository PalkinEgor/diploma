```python
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
with open('D:\\учеба\\diploma\\data\\training_results.json', 'r') as f:
    data = json.load(f)
```

```python
data_df = pd.DataFrame(data)
data_df.head(5)
```

```python
accuracies = []
words_len = []
for item in data:
    words_len.append(len(item['text'].split(' ')) if len(item['text'].split(' ')) <= 75 else 75)
    accuracies.append(item['accuracy'])
```

```python
plt.hist(accuracies, bins=50)
plt.xlabel('Значения')
plt.ylabel('Частота')
plt.title('Гистограмма accuracy')
plt.show()
```

```python
plt.hist(words_len, bins=50)
plt.xlabel('Значения')
plt.ylabel('Частота')
plt.title('Гистограмма длин предложений')
plt.show()
```

```python
plt.scatter(words_len, accuracies)
plt.xlabel('Длина предложения')
plt.ylabel('Accuracy')
plt.title('Зависимость accuracy от длины предложения')
plt.show()
```
